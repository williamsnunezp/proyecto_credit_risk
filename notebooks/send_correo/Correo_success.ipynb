```python
# 2. OBTENER PARÁMETROS
send_to = dbutils.widgets.get("send_to")
mensaje = dbutils.widgets.get("mensaje")
```

```python
# 3. CREAR MENSAJE

msg = EmailMessage()
correo_origen = 'xxxx@gmail.com'

msg['Subject'] = 'Pipeline ETL Ejecutado Exitosamente, ya puede consumir sus tablas desde Databricks'
msg['From'] = correo_origen
msg['To'] = send_to

msg.set_content(mensaje)

# 4. DATOS DEL SERVIDOR SMTP DE GMAIL

smtp_server = 'smtp.gmail.com'
smtp_port = 587
clave_app = 'xxxxx'  # 

# 5. ENVIAR EL CORREO
try:
    with smtplib.SMTP(smtp_server, smtp_port) as smtp:
        smtp.ehlo()
        smtp.starttls()
        smtp.login(correo_origen, clave_app)
        smtp.send_message(msg)
        smtp.quit()
        print(' Correo enviado con éxito.')
        
except Exception as e:
    print(f' Error al enviar el correo: {e}')
```
