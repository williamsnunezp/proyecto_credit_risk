```python
# 2. OBTENER PARÁMETROS
send_to = dbutils.widgets.get("send_to")
mensaje = dbutils.widgets.get("mensaje")
```

```python
# 3. CONFIGURACIÓN Y ENVÍO DEL CORREO

# -- Parámetros del correo --
sender_email = "xxxx@gmail.com"
sender_password = "xxxxx"
subject = (
    "Pipeline ETL Ejecutado Exitosamente: "
    "Tablas listas para consumir en Databricks"
)

# -- Creación del mensaje --
msg = EmailMessage()
msg["Subject"] = subject
msg["From"] = sender_email
msg["To"] = send_to
msg.set_content(mensaje)

# -- Envío del correo --
smtp_server = "smtp.gmail.com"
smtp_port = 587

try:
    with smtplib.SMTP(smtp_server, smtp_port) as server:
        server.starttls()
        server.login(sender_email, sender_password)
        server.send_message(msg)
    print("Correo enviado con éxito.")
except Exception as e:
    print(f"Error al enviar el correo: {e}")
```
